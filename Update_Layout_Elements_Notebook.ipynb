```python
# TASK - SET VARIABLES - OLD TEXT VARIABLES ARE STANDARD 
# YOUR TASK - IF YOU CHOOSE TO ACCEPT IT IS TO SET THE NEW VARIABLES:
    # unit_name, location_text, and region_text with the appropriate text

unit_temp_txt = "(XXX_UNIT NAME_XXX)"
unit_name = "Cape Cod National Seashore"

location_temp_text = "(XXXX_Location_XXX)"
location_text = "Township of Truro"

region_temp_text = "Washington Support Office"
region_text = "Northeast Region"

# TASK - Set up your Projection Variable
# If you don't know what projection you need to use - find the most current projection here:
    # https://spatialreference.org/ref/?search=connecticut
    
target_projection = "NAD_1983_2011_StatePlane_Connecticut_FIPS_0600_Ft_US"  # Example: WKID for WGS 1984
```

```python
# TASK - UPDATE TEXT ELEMENTS
# Reference https://pro.arcgis.com/en/pro-app/latest/arcpy/mapping/textelement-class.htm

import arcpy

# REPLACE UNIT NAME

def replace_unit_in_layouts(unit_temp_txt, unit_name):
    try:
        # Open the ArcGIS Pro project
        aprx = arcpy.mp.ArcGISProject("CURRENT")

        # Loop through all layouts in the project
        for layout in aprx.listLayouts():
            # Loop through all text elements in the layout
            for text_element in layout.listElements('TEXT_ELEMENT'):
                if unit_temp_txt in text_element.text:
                    text_element.text = text_element.text.replace(unit_temp_txt, unit_name)

        # Save changes to the project
        aprx.save()

    except Exception as e:
        print("An error occurred:", e)

replace_unit_in_layouts(unit_temp_txt, unit_name)

#REPLACE LOCATION TEXT

def replace_location_in_layouts(location_temp_text, location_text):
    try:
        # Open the ArcGIS Pro project
        aprx = arcpy.mp.ArcGISProject("CURRENT")

        # Loop through all layouts in the project
        for layout in aprx.listLayouts():
            # Loop through all text elements in the layout
            for text_element in layout.listElements('TEXT_ELEMENT'):
                if location_temp_text in text_element.text:
                    text_element.text = text_element.text.replace(location_temp_text, location_text)

        # Save changes to the project
        aprx.save()

    except Exception as e:
        print("An error occurred:", e)

replace_location_in_layouts(location_temp_text, location_text)

# REPLACE REGION TEXT
def replace_region_in_layouts(region_temp_text, region_text):
    try:
        # Open the ArcGIS Pro project
        aprx = arcpy.mp.ArcGISProject("CURRENT")

        # Loop through all layouts in the project
        for layout in aprx.listLayouts():
            # Loop through all text elements in the layout
            for text_element in layout.listElements('TEXT_ELEMENT'):
                if region_temp_text in text_element.text:
                    text_element.text = text_element.text.replace(region_temp_text, region_text)

        # Save changes to the project
        aprx.save()

    except Exception as e:
        print("An error occurred:", e)

replace_region_in_layouts(region_temp_text, region_text)
```

```python
# TASK - SET THE MAP PROJECTIONS - You will not need to set any variables here - 
    # all variables were set in cell 1 above

import arcpy

def set_projection_for_maps(target_projection):
    try:
        # Open the ArcGIS Pro project
        aprx = arcpy.mp.ArcGISProject("CURRENT")

        # Loop through all maps in the project
        for map in aprx.listMaps():
            # Set the projection for the map
            map.spatialReference = arcpy.SpatialReference(target_projection)

        # Save changes to the project
        aprx.save()

    except Exception as e:
        print("An error occurred:", e)

set_projection_for_maps(target_projection)
```

```python
#TASK - Update Legend items in all layouts and set Legend to show only 
    #Layers visible in extent
    
import arcpy

def update_legend_in_layouts():
    try:
        # Open the ArcGIS Pro project
        aprx = arcpy.mp.ArcGISProject("CURRENT")

        # Get the active map
        map = aprx.activeMap

        # Loop through all layouts in the project
        for layout in aprx.listLayouts():
            # Get the legend element in the layout
            legend_element = layout.listElements('LEGEND_ELEMENT')[0]

            # Clear existing legend items
            legend_items = legend_element.items[:]
            for legend_item in legend_items:
                legend_element.removeItem(legend_item)

            # Add legend items based on current map extent
            for layer in map.listLayers():
                if layer.visible:
                    legend_element.addItem(layer)

        # Save changes to the project
        aprx.save()

    except Exception as e:
        print("An error occurred:", e)

# This line runs the function defined above
update_legend_in_layouts()
```
